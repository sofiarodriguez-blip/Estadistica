```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
from math import sqrt

# Problema 21: Prueba de Hipótesis para el Peso Promedio de Cereal

# 1. Plantear las hipótesis nula y alternativa.
# H₀: μ = 500 g (La máquina está calibrada)
# H₁: μ < 500 g (La máquina está subllenando - Prueba unilateral izquierda)
print("1. Hipótesis:")
print("   H₀: μ = 500 g")
print("   H₁: μ < 500 g (Prueba unilateral izquierda)")

# 2. Establecer el nivel de significancia α.
alpha = 0.05
print(f"\n2. Nivel de significancia α = {alpha}")

# Definir variables del problema
media_muestral = 495
media_poblacional_h0 = 500
desviacion_estandar_poblacional = 15
tamano_muestra = 40

# Calcular el error estándar
error_estandar = desviacion_estandar_poblacional / sqrt(tamano_muestra)

# 3. Identificar el estadístico de prueba y su distribución.
# Se usa el estadístico Z porque la desviación estándar poblacional es conocida y n > 30.
# Distribución: Normal Estándar.

# 4. Determinar el valor crítico y la región de rechazo.
# Para una prueba unilateral izquierda con α = 0.05
z_critical = norm.ppf(alpha)
print(f"\n4. Valor crítico de Z: {z_critical:.4f}")
print(f"   Región de rechazo: Z <= {z_critical:.4f}")

# 5. Calcule el estadístico de prueba y el valor p.
# Calcular el estadístico de prueba Z
z_estadistico = (media_muestral - media_poblacional_h0) / error_estandar
print(f"\n5. Estadístico de prueba Z calculado: {z_estadistico:.4f}")

# Calcular el valor p para una prueba unilateral izquierda
valor_p = norm.cdf(z_estadistico)
print(f"   Valor p: {valor_p:.4f}")

# 6. Graficar la distribución correspondiente, indicando claramente la región de rechazo, el valor crítico y el estadístico calculado.
# Valores del eje x para la gráfica
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1) # Densidad de probabilidad de la distribución normal estándar

# Crear la figura y los ejes
fig, ax = plt.subplots(figsize=(10, 6))

# Graficar la curva de densidad de probabilidad
ax.plot(x, y, label='Distribución Normal Estándar')

# Rellenar la región de rechazo (cola izquierda)
x_reject = np.linspace(-4, z_critical, 100)
y_reject = norm.pdf(x_reject, 0, 1)
ax.fill_between(x_reject, y_reject, color='red', alpha=0.5, label=f'Región de Rechazo (α = {alpha})')

# Dibujar líneas para el valor crítico y el estadístico de prueba
ax.axvline(z_critical, color='red', linestyle='--', label=f'Valor Crítico Z = {z_critical:.4f}')
ax.axvline(z_estadistico, color='blue', linestyle='-', label=f'Estadístico Z = {z_estadistico:.4f}')

# Añadir texto para el valor p
# Posicionar el texto cerca del estadístico de prueba
text_x = z_estadistico
text_y = norm.pdf(z_estadistico, 0, 1) + 0.02 # Un poco por encima de la curva
ax.text(text_x, text_y, f'Valor p = {valor_p:.4f}', color='blue', ha='center')


# Añadir etiquetas y título
ax.set_xlabel('Valores Z')
ax.set_ylabel('Densidad de Probabilidad')
ax.set_title('Distribución Normal Estándar con Región de Rechazo y Estadístico de Prueba')
ax.legend()
ax.grid(True)
plt.show()

# 7. Tome una decisión estadística y concluya en términos prácticos, explicando el significado del valor p.
print("\n7. Decisión Estadística y Conclusión:")
if valor_p <= alpha:
    decision_estadistica = "Rechazar H0"
    conclusion_practica = f"Dado que el valor p ({valor_p:.4f}) es menor que el nivel de significancia α ({alpha}), rechazamos la hipótesis nula."
    conclusion_practica += "\nEn términos prácticos: Hay suficiente evidencia estadística, al nivel de significancia de 0.05, para concluir que el peso promedio de los paquetes de cereal es significativamente menor a 500 g. Esto sugiere que la máquina de llenado probablemente está subcalibrada."
else:
    decision_estadistica = "No rechazar H0"
    conclusion_practica = f"Dado que el valor p ({valor_p:.4f}) es mayor que el nivel de significancia α ({alpha}), no rechazamos la hipótesis nula."
    conclusion_practica += "\nEn términos prácticos: No hay suficiente evidencia estadística, al nivel de significancia de 0.05, para concluir que el peso promedio de los paquetes de cereal es significativamente diferente de 500 g. No podemos afirmar con esta evidencia que la máquina esté subllenando."

print(f"   Decisión Estadística: {decision_estadistica}")
print(f"   Conclusión Práctica: {conclusion_practica}")
print(f"\n   Significado del Valor p:")
print(f"   El valor p de {valor_p:.4f} es la probabilidad de obtener una media muestral tan baja como 495 g (o más baja) si la verdadera media poblacional fuera realmente 500 g. Un valor p pequeño indica que el resultado observado es improbable bajo la hipótesis nula, lo que nos lleva a rechazarla.")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
from math import sqrt

# Problema de Prueba de Hipótesis: Promedio de la Clase vs Promedio Nacional

# Un profesor afirma que el promedio de su clase en un examen nacional es diferente al promedio
# nacional de 75 puntos. Una muestra de 36 estudiantes de su clase tiene una media de 78. La
# desviación estándar poblacional es σ = 12. Con un nivel de significancia: α = 0,05.

# 1. Plantear las hipótesis nula y alternativa.
# H₀: μ = 75 puntos (El promedio de la clase es igual al promedio nacional)
# H₁: μ ≠ 75 puntos (El promedio de la clase es diferente al promedio nacional - Prueba bilateral)
print("1. Hipótesis:")
print("   H₀: μ = 75 puntos")
print("   H₁: μ ≠ 75 puntos (Prueba bilateral)")

# 2. Establecer el nivel de significancia α.
alpha = 0.05
print(f"\n2. Nivel de significancia α = {alpha}")

# Definir variables del problema
media_muestral = 78
media_poblacional_h0 = 75
desviacion_estandar_poblacional = 12
tamano_muestra = 36

# Calcular el error estándar
error_estandar = desviacion_estandar_poblacional / sqrt(tamano_muestra)

# 3. Identificar el estadístico de prueba y su distribución.
# Se usa el estadístico Z porque la desviación estándar poblacional es conocida y n >= 30.
# Distribución: Normal Estándar.

# 4. Determinar los valores críticos y la región de rechazo.
# Para una prueba bilateral con α = 0.05, el área de rechazo se divide en dos colas (α/2 en cada cola).
alpha_half = alpha / 2
z_critical_lower = norm.ppf(alpha_half)
z_critical_upper = norm.ppf(1 - alpha_half)
print(f"\n4. Valores críticos de Z: {z_critical_lower:.4f} y {z_critical_upper:.4f}")
print(f"   Región de rechazo: Z <= {z_critical_lower:.4f} o Z >= {z_critical_upper:.4f}")

# 5. Calcule el estadístico de prueba y el valor p.
# Calcular el estadístico de prueba Z
z_estadistico = (media_muestral - media_poblacional_h0) / error_estandar
print(f"\n5. Estadístico de prueba Z calculado: {z_estadistico:.4f}")

# Calcular el valor p para una prueba bilateral
# Para una prueba bilateral, el valor p es el doble del área en la cola más allá del estadístico de prueba.
# Como nuestro estadístico Z es positivo, calculamos el área en la cola derecha y la multiplicamos por 2.
valor_p = 2 * (1 - norm.cdf(abs(z_estadistico))) # Usamos abs() para manejar estadísticos Z negativos también
print(f"   Valor p: {valor_p:.4f}")


# 6. Graficar la distribución correspondiente, indicando claramente la región de rechazo, los valores críticos y el estadístico calculado.
# Valores del eje x para la gráfica
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1) # Densidad de probabilidad de la distribución normal estándar

# Crear la figura y los ejes
fig, ax = plt.subplots(figsize=(10, 6))

# Graficar la curva de densidad de probabilidad
ax.plot(x, y, label='Distribución Normal Estándar')

# Rellenar la región de rechazo (dos colas)
x_reject_lower = np.linspace(-4, z_critical_lower, 100)
y_reject_lower = norm.pdf(x_reject_lower, 0, 1)
ax.fill_between(x_reject_lower, y_reject_lower, color='red', alpha=0.5, label=f'Región de Rechazo (α/2 = {alpha_half})')

x_reject_upper = np.linspace(z_critical_upper, 4, 100)
y_reject_upper = norm.pdf(x_reject_upper, 0, 1)
ax.fill_between(x_reject_upper, y_reject_upper, color='red', alpha=0.5) # No repetir la etiqueta

# Dibujar líneas para los valores críticos y el estadístico de prueba
ax.axvline(z_critical_lower, color='red', linestyle='--', label=f'Valores Críticos Z = {z_critical_lower:.4f}, {z_critical_upper:.4f}')
ax.axvline(z_critical_upper, color='red', linestyle='--')
ax.axvline(z_estadistico, color='blue', linestyle='-', label=f'Estadístico Z = {z_estadistico:.4f}')

# Añadir texto para el valor p
# Posicionar el texto cerca del estadístico de prueba
text_x = z_estadistico
text_y = norm.pdf(z_estadistico, 0, 1) + 0.02 # Un poco por encima de la curva
ax.text(text_x, text_y, f'Valor p/2 = {valor_p/2:.4f}', color='blue', ha='center') # Mostrar p/2 en la gráfica bilateral
ax.text(0, ax.get_ylim()[1]*0.9, f'Valor p total = {valor_p:.4f}', color='purple', ha='center') # Mostrar valor p total en el centro


# Añadir etiquetas y título
ax.set_xlabel('Valores Z')
ax.set_ylabel('Densidad de Probabilidad')
ax.set_title('Distribución Normal Estándar con Regiones de Rechazo y Estadístico de Prueba (Bilateral)')
ax.legend()
ax.grid(True)
plt.show()

# 7. Tome una decisión estadística y concluya en términos prácticos, explicando el significado del valor p.
print("\n7. Decisión Estadística y Conclusión:")
if valor_p <= alpha:
    decision_estadistica = "Rechazar H0"
    conclusion_practica = f"Dado que el valor p ({valor_p:.4f}) es menor que el nivel de significancia α ({alpha}), rechazamos la hipótesis nula."
    conclusion_practica += "\nEn términos prácticos: Hay suficiente evidencia estadística, al nivel de significancia de 0.05, para concluir que el promedio de la clase en el examen nacional es significativamente diferente de 75 puntos."
else:
    decision_estadistica = "No rechazar H0"
    conclusion_practica = f"Dado que el valor p ({valor_p:.4f}) es mayor que el nivel de significancia α ({alpha}), no rechazamos la hipótesis nula."
    conclusion_practica += "\nEn términos prácticos: No hay suficiente evidencia estadística, al nivel de significancia de 0.05, para concluir que el promedio de la clase en el examen nacional es significativamente diferente de 75 puntos. No podemos afirmar con esta evidencia que el promedio de la clase sea distinto al promedio nacional."

print(f"   Decisión Estadística: {decision_estadistica}")
print(f"   Conclusión Práctica: {conclusion_practica}")
print(f"\n   Significado del Valor p:")
print(f"   El valor p de {valor_p:.4f} es la probabilidad de obtener una media muestral tan extrema como 78 (o 72, debido a la naturaleza bilateral) o más extrema, si la verdadera media poblacional fuera realmente 75. Un valor p pequeño (< α) indica que el resultado observado es improbable bajo la hipótesis nula, lo que nos lleva a rechazarla.")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
from math import sqrt

# Problema de Prueba de Hipótesis: Reducción de Presión Arterial

# Un laboratorio afirma que su medicamento reduce la presión arterial en 10 mmHg en promedio. Un
# estudio con 25 pacientes muestra una reducción promedio de 8 mmHg, con σ = 3 mmHg. Con un
# nivel de significancia α = 0,05.

# 1. Plantear las hipótesis nula y alternativa.
# H₀: μ = 10 mmHg (La reducción promedio es 10 mmHg)
# H₁: μ < 10 mmHg (La reducción promedio es menor a 10 mmHg - Prueba unilateral izquierda)
print("1. Hipótesis:")
print("   H₀: μ = 10 mmHg")
print("   H₁: μ < 10 mmHg (Prueba unilateral izquierda)")

# 2. Establecer el nivel de significancia α.
alpha = 0.05
print(f"\n2. Nivel de significancia α = {alpha}")

# Definir variables del problema
media_muestral = 8
media_poblacional_h0 = 10
desviacion_estandar_poblacional = 3
tamano_muestra = 25

# Calcular el error estándar
error_estandar = desviacion_estandar_poblacional / sqrt(tamano_muestra)

# 3. Identificar el estadístico de prueba y su distribución.
# Se usa el estadístico Z porque la desviación estándar poblacional es conocida.
# Distribución: Normal Estándar.

# 4. Determinar el valor crítico y la región de rechazo.
# Para una prueba unilateral izquierda con α = 0.05
z_critical = norm.ppf(alpha)
print(f"\n4. Valor crítico de Z: {z_critical:.4f}")
print(f"   Región de rechazo: Z <= {z_critical:.4f}")

# 5. Calcule el estadístico de prueba y el valor p.
# Calcular el estadístico de prueba Z
z_estadistico = (media_muestral - media_poblacional_h0) / error_estandar
print(f"\n5. Estadístico de prueba Z calculado: {z_estadistico:.4f}")

# Calcular el valor p para una prueba unilateral izquierda
valor_p = norm.cdf(z_estadistico)
print(f"   Valor p: {valor_p:.4f}")

# 6. Graficar la distribución correspondiente, indicando claramente la región de rechazo, el valor crítico y el estadístico calculado.
# Valores del eje x para la gráfica
x = np.linspace(-5, 1, 1000) # Ajustar rango para visualizar mejor el estadístico negativo
y = norm.pdf(x, 0, 1) # Densidad de probabilidad de la distribución normal estándar

# Crear la figura y los ejes
fig, ax = plt.subplots(figsize=(10, 6))

# Graficar la curva de densidad de probabilidad
ax.plot(x, y, label='Distribución Normal Estándar')

# Rellenar la región de rechazo (cola izquierda)
x_reject = np.linspace(-5, z_critical, 100)
y_reject = norm.pdf(x_reject, 0, 1)
ax.fill_between(x_reject, y_reject, color='red', alpha=0.5, label=f'Región de Rechazo (α = {alpha})')

# Dibujar líneas para el valor crítico y el estadístico de prueba
ax.axvline(z_critical, color='red', linestyle='--', label=f'Valor Crítico Z = {z_critical:.4f}')
ax.axvline(z_estadistico, color='blue', linestyle='-', label=f'Estadístico Z = {z_estadistico:.4f}')

# Añadir texto para el valor p
# Posicionar el texto cerca del estadístico de prueba
text_x = z_estadistico
text_y = norm.pdf(z_estadistico, 0, 1) + 0.02 # Un poco por encima de la curva
ax.text(text_x, text_y, f'Valor p = {valor_p:.4f}', color='blue', ha='center')


# Añadir etiquetas y título
ax.set_xlabel('Valores Z')
ax.set_ylabel('Densidad de Probabilidad')
ax.set_title('Distribución Normal Estándar con Región de Rechazo y Estadístico de Prueba')
ax.legend()
ax.grid(True)
plt.show()

# 7. Tome una decisión estadística y concluya en términos prácticos, explicando el significado del valor p.
print("\n7. Decisión Estadística y Conclusión:")
if valor_p <= alpha:
    decision_estadistica = "Rechazar H0"
    conclusion_practica = f"Dado que el valor p ({valor_p:.4f}) es menor que el nivel de significancia α ({alpha}), rechazamos la hipótesis nula."
    conclusion_practica += "\nEn términos prácticos: Hay suficiente evidencia estadística, al nivel de significancia de 0.05, para concluir que la verdadera reducción promedio de la presión arterial con este medicamento es significativamente menor a 10 mmHg."
else:
    decision_estadistica = "No rechazar H0"
    conclusion_practica = f"Dado que el valor p ({valor_p:.4f}) es mayor que el nivel de significancia α ({alpha}), no rechazamos la hipótesis nula."
    conclusion_practica += "\nEn términos prácticos: No hay suficiente evidencia estadística, al nivel de significancia de 0.05, para concluir que la verdadera reducción promedio de la presión arterial con este medicamento es significativamente diferente de 10 mmHg. No podemos afirmar con esta evidencia que la reducción sea menor a 10 mmHg."


print(f"   Decisión Estadística: {decision_estadistica}")
print(f"   Conclusión Práctica: {conclusion_practica}")
print(f"\n   Significado del Valor p:")
print(f"   El valor p de {valor_p:.4f} es la probabilidad de obtener una reducción promedio tan baja como 8 mmHg (o más baja) si la verdadera reducción promedio poblacional fuera realmente 10 mmHg. Un valor p pequeño (< α) indica que el resultado observado es improbable bajo la hipótesis nula, lo que nos lleva a rechazarla.")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import chi2

# Problema de Prueba de Hipótesis: Varianza del Tiempo de Duración de Llamadas

# Una compañía de telefonía afirma que la varianza del tiempo de duración de las llamadas es de
# σ² = 2,5 minutos². Se toma una muestra de n = 16 llamadas y se obtiene una varianza muestral
# de s² = 4. ¿Hay evidencia al 5 % de significancia de que la varianza poblacional difiere de 2.5?

# 1. Plantear las hipótesis nula y alternativa.
# H₀: σ² = 2.5 (La varianza poblacional es 2.5)
# H₁: σ² ≠ 2.5 (La varianza poblacional difiere de 2.5 - Prueba bilateral)
print("1. Hipótesis:")
print("   H₀: σ² = 2.5 minutos²")
print("   H₁: σ² ≠ 2.5 minutos² (Prueba bilateral)")

# 2. Establecer el nivel de significancia α.
alpha = 0.05
print(f"\n2. Nivel de significancia α = {alpha}")

# Definir variables del problema
varianza_poblacional_h0 = 2.5
varianza_muestral = 4
tamano_muestra = 16

# 3. Identificar el estadístico de prueba y su distribución.
# Se usa el estadístico Chi-cuadrado (χ²) para probar una hipótesis sobre una única varianza poblacional.
# Distribución: Chi-cuadrado con n-1 grados de libertad.

# Calcular los grados de libertad
grados_libertad = tamano_muestra - 1
print(f"\n3. Estadístico de prueba: χ²")
print(f"   Distribución: Chi-cuadrado con {grados_libertad} grados de libertad")


# 4. Determinar los valores críticos y la región de rechazo.
# Para una prueba bilateral con α = 0.05, el área de rechazo se divide en dos colas (α/2 en cada cola).
alpha_half = alpha / 2
chi2_critical_lower = chi2.ppf(alpha_half, grados_libertad)
chi2_critical_upper = chi2.ppf(1 - alpha_half, grados_libertad)
print(f"\n4. Valores críticos de χ²: {chi2_critical_lower:.4f} y {chi2_critical_upper:.4f}")
print(f"   Región de rechazo: χ² <= {chi2_critical_lower:.4f} o χ² >= {chi2_critical_upper:.4f}")

# 5. Calcule el estadístico de prueba y el valor p.
# Calcular el estadístico de prueba Chi-cuadrado (χ²)
# Fórmula: χ² = ((n - 1) * s²) / σ²
chi2_estadistico = ((tamano_muestra - 1) * varianza_muestral) / varianza_poblacional_h0
print(f"\n5. Estadístico de prueba χ² calculado: {chi2_estadistico:.4f}")

# Calcular el valor p para una prueba bilateral de varianza
# Para una prueba bilateral, el valor p es el doble del área en la cola más allá del estadístico de prueba.
# Si el estadístico es mayor que la media de la distribución Chi-cuadrado (que es igual a los grados de libertad),
# calculamos el área en la cola superior y multiplicamos por 2.
# Si es menor, calculamos el área en la cola inferior y multiplicamos por 2.
# En la distribución Chi-cuadrado, la media es los grados de libertad. El estadístico es 24, gl es 15.
# Calculamos el área en la cola derecha (1 - cdf).
if chi2_estadistico > grados_libertad:
    valor_p = 2 * (1 - chi2.cdf(chi2_estadistico, grados_libertad))
else:
    valor_p = 2 * chi2.cdf(chi2_estadistico, grados_libertad)

# Nota: Una forma más general y precisa para pruebas bilaterales es tomar el mínimo del área en la cola izquierda o derecha
# y multiplicarlo por 2.
valor_p_alt = 2 * min(chi2.cdf(chi2_estadistico, grados_libertad), 1 - chi2.cdf(chi2_estadistico, grados_libertad))

print(f"   Valor p: {valor_p_alt:.4f}")


# 6. Graficar la distribución correspondiente, indicando claramente la región de rechazo, los valores críticos y el estadístico calculado.
# Valores del eje x para la gráfica de la distribución Chi-cuadrado
# El rango debe ir desde 0 hasta un valor que incluya los valores críticos y el estadístico.
x = np.linspace(0, chi2.ppf(0.999, grados_libertad), 1000)
y = chi2.pdf(x, grados_libertad) # Densidad de probabilidad de la distribución Chi-cuadrado

# Crear la figura y los ejes
fig, ax = plt.subplots(figsize=(10, 6))

# Graficar la curva de densidad de probabilidad
ax.plot(x, y, label=f'Distribución χ² ({grados_libertad} gl)')

# Rellenar la región de rechazo (dos colas)
x_reject_lower = np.linspace(0, chi2_critical_lower, 100)
y_reject_lower = chi2.pdf(x_reject_lower, grados_libertad)
ax.fill_between(x_reject_lower, y_reject_lower, color='red', alpha=0.5, label=f'Región de Rechazo (α/2 = {alpha_half})')

x_reject_upper = np.linspace(chi2_critical_upper, x[-1], 100)
y_reject_upper = chi2.pdf(x_reject_upper, grados_libertad)
ax.fill_between(x_reject_upper, y_reject_upper, color='red', alpha=0.5) # No repetir la etiqueta

# Dibujar líneas para los valores críticos y el estadístico de prueba
ax.axvline(chi2_critical_lower, color='red', linestyle='--', label=f'Valores Críticos χ² = {chi2_critical_lower:.4f}, {chi2_critical_upper:.4f}')
ax.axvline(chi2_critical_upper, color='red', linestyle='--')
ax.axvline(chi2_estadistico, color='blue', linestyle='-', label=f'Estadístico χ² = {chi2_estadistico:.4f}')

# Añadir texto para el valor p
# Posicionar el texto cerca del estadístico de prueba
# Para prueba bilateral, el valor p se refiere al área total en ambas colas.
# Podemos indicar el área de una cola y el valor p total.
area_cola_estadistico = 1 - chi2.cdf(chi2_estadistico, grados_libertad) # Área en la cola derecha (para este caso particular)
text_x = chi2_estadistico
text_y = chi2.pdf(chi2_estadistico, grados_libertad) + 0.005 # Un poco por encima de la curva
ax.text(text_x, text_y, f'Área cola = {area_cola_estadistico:.4f}', color='blue', ha='center')
ax.text(grados_libertad, ax.get_ylim()[1]*0.9, f'Valor p total = {valor_p_alt:.4f}', color='purple', ha='center') # Mostrar valor p total cerca de la media


# Añadir etiquetas y título
ax.set_xlabel('Valores χ²')
ax.set_ylabel('Densidad de Probabilidad')
ax.set_title(f'Distribución Chi-cuadrado con {grados_libertad} gl con Regiones de Rechazo y Estadístico de Prueba (Bilateral)')
ax.legend()
ax.grid(True)
plt.show()

# 7. Tome una decisión estadística y concluya en términos prácticos, explicando el significado del valor p.
print("\n7. Decisión Estadística y Conclusión:")
# Criterio del Valor Crítico
if chi2_estadistico <= chi2_critical_lower or chi2_estadistico >= chi2_critical_upper:
    decision_estadistica = "Rechazar H0 (usando valores críticos)"
    conclusion_critica = f"Dado que el estadístico de prueba χ² ({chi2_estadistico:.4f}) cae en la región de rechazo (menor o igual a {chi2_critical_lower:.4f} o mayor o igual a {chi2_critical_upper:.4f}), rechazamos la hipótesis nula."
else:
    decision_estadistica = "No rechazar H0 (usando valores críticos)"
    conclusion_critica = f"Dado que el estadístico de prueba χ² ({chi2_estadistico:.4f}) no cae en la región de rechazo ({chi2_critical_lower:.4f} < χ² < {chi2_critical_upper:.4f}), no rechazamos la hipótesis nula."

# Criterio del Valor p
if valor_p_alt <= alpha:
     decision_p_valor = "Rechazar H0 (usando valor p)"
     conclusion_p_valor = f"Dado que el valor p ({valor_p_alt:.4f}) es menor que el nivel de significancia α ({alpha}), rechazamos la hipótesis nula."
else:
     decision_p_valor = "No rechazar H0 (usando valor p)"
     conclusion_p_valor = f"Dado que el valor p ({valor_p_alt:.4f}) es mayor que el nivel de significancia α ({alpha}), no rechazamos la hipótesis nula."

print(f"   Decisión Estadística (Criterio del Valor Crítico): {decision_estadistica}")
print(f"   Conclusión (Criterio del Valor Crítico): {conclusion_critica}")
print(f"   Decisión Estadística (Criterio del Valor p): {decision_p_valor}")
print(f"   Conclusión (Criterio del Valor p): {conclusion_p_valor}")


print(f"\n   Conclusión General en Términos Prácticos:")
if valor_p_alt <= alpha:
     print(f"   Al nivel de significancia de {alpha*100}%, hay suficiente evidencia estadística para concluir que la verdadera varianza del tiempo de duración de las llamadas es significativamente diferente de 2.5 minutos².")
else:
     print(f"   Al nivel de significancia de {alpha*100}%, no hay suficiente evidencia estadística para concluir que la verdadera varianza del tiempo de duración de las llamadas es significativamente diferente de 2.5 minutos². No podemos afirmar con esta evidencia que la varianza poblacional difiere de 2.5.")

print(f"\n   Significado del Valor p:")
print(f"   El valor p de {valor_p_alt:.4f} es la probabilidad de obtener una varianza muestral tan extrema como 4 (o tan baja como para resultar en un estadístico χ² tan extremo) si la verdadera varianza poblacional fuera realmente 2.5 minutos². Un valor p pequeño (< α) indica que el resultado observado es improbable bajo la hipótesis nula, lo que nos lleva a rechazarla.")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import chi2

# Problema de Prueba de Hipótesis: Varianza del Grosor de Piezas Metálicas

# El control de calidad de una fábrica sostiene que la varianza en el grosor de piezas metálicas es de
# al menos σ² = 0,5 mm². Una muestra de n = 12 piezas dio una varianza s² = 0,3. ¿Se puede
# concluir al 5 % que la varianza poblacional es menor que la garantizada?

# 1. Plantear las hipótesis nula y alternativa.
# H₀: σ² >= 0.5 mm² (La varianza es al menos 0.5 - Para la prueba, usamos σ² = 0.5)
# H₁: σ² < 0.5 mm² (La varianza es menor que 0.5 - Prueba unilateral izquierda)
print("1. Hipótesis:")
print("   H₀: σ² = 0.5 mm² (asumimos igualdad para la prueba)")
print("   H₁: σ² < 0.5 mm² (Prueba unilateral izquierda)")

# 2. Establecer el nivel de significancia α.
alpha = 0.05
print(f"\n2. Nivel de significancia α = {alpha}")

# Definir variables del problema
varianza_poblacional_h0 = 0.5
varianza_muestral = 0.3
tamano_muestra = 12

# 3. Identificar el estadístico de prueba y su distribución.
# Se usa el estadístico Chi-cuadrado (χ²) para probar una hipótesis sobre una única varianza poblacional.
# Distribución: Chi-cuadrado con n-1 grados de libertad.

# Calcular los grados de libertad
grados_libertad = tamano_muestra - 1
print(f"\n3. Estadístico de prueba: χ²")
print(f"   Distribución: Chi-cuadrado con {grados_libertad} grados de libertad")


# 4. Determinar el valor crítico y la región de rechazo.
# Para una prueba unilateral izquierda con α = 0.05 y {grados_libertad} grados de libertad
chi2_critical = chi2.ppf(alpha, grados_libertad)
print(f"\n4. Valor crítico de χ²: {chi2_critical:.4f}")
print(f"   Región de rechazo: χ² <= {chi2_critical:.4f}")

# 5. Calcule el estadístico de prueba y el valor p.
# Calcular el estadístico de prueba Chi-cuadrado (χ²)
# Fórmula: χ² = ((n - 1) * s²) / σ²
chi2_estadistico = ((tamano_muestra - 1) * varianza_muestral) / varianza_poblacional_h0
print(f"\n5. Estadístico de prueba χ² calculado: {chi2_estadistico:.4f}")

# Calcular el valor p para una prueba unilateral izquierda
valor_p = chi2.cdf(chi2_estadistico, grados_libertad)
print(f"   Valor p: {valor_p:.4f}")


# 6. Graficar la distribución correspondiente, indicando claramente la región de rechazo, el valor crítico y el estadístico calculado.
# Valores del eje x para la gráfica de la distribución Chi-cuadrado
# El rango debe ir desde 0 hasta un valor que incluya el valor crítico y el estadístico.
x_max = max(chi2.ppf(0.999, grados_libertad), chi2_critical, chi2_estadistico) + 2 # Añadir un margen
x = np.linspace(0, x_max, 1000)
y = chi2.pdf(x, grados_libertad) # Densidad de probabilidad de la distribución Chi-cuadrado

# Crear la figura y los ejes
fig, ax = plt.subplots(figsize=(10, 6))

# Graficar la curva de densidad de probabilidad
ax.plot(x, y, label=f'Distribución χ² ({grados_libertad} gl)')

# Rellenar la región de rechazo (cola izquierda)
x_reject = np.linspace(0, chi2_critical, 100)
y_reject = chi2.pdf(x_reject, grados_libertad)
ax.fill_between(x_reject, y_reject, color='red', alpha=0.5, label=f'Región de Rechazo (α = {alpha})')

# Dibujar líneas para el valor crítico y el estadístico de prueba
ax.axvline(chi2_critical, color='red', linestyle='--', label=f'Valor Crítico χ² = {chi2_critical:.4f}')
ax.axvline(chi2_estadistico, color='blue', linestyle='-', label=f'Estadístico χ² = {chi2_estadistico:.4f}')

# Añadir texto para el valor p
# Posicionar el texto cerca del estadístico de prueba
text_x = chi2_estadistico
text_y = chi2.pdf(chi2_estadistico, grados_libertad) + 0.005 # Un poco por encima de la curva
ax.text(text_x, text_y, f'Valor p = {valor_p:.4f}', color='blue', ha='center')


# Añadir etiquetas y título
ax.set_xlabel('Valores χ²')
ax.set_ylabel('Densidad de Probabilidad')
ax.set_title(f'Distribución Chi-cuadrado con {grados_libertad} gl con Región de Rechazo y Estadístico de Prueba (Unilateral Izquierda)')
ax.legend()
ax.grid(True)
plt.show()

# 7. Tome una decisión estadística y concluya en términos prácticos, explicando el significado del valor p.
print("\n7. Decisión Estadística y Conclusión:")
# Criterio del Valor Crítico
if chi2_estadistico <= chi2_critical:
    decision_estadistica = "Rechazar H0 (usando valor crítico)"
    conclusion_critica = f"Dado que el estadístico de prueba χ² ({chi2_estadistico:.4f}) cae en la región de rechazo (menor o igual a {chi2_critical:.4f}), rechazamos la hipótesis nula."
else:
    decision_estadistica = "No rechazar H0 (usando valor crítico)"
    conclusion_critica = f"Dado que el estadístico de prueba χ² ({chi2_estadistico:.4f}) no cae en la región de rechazo (mayor a {chi2_critical:.4f}), no rechazamos la hipótesis nula."

# Criterio del Valor p
if valor_p <= alpha:
     decision_p_valor = "Rechazar H0 (usando valor p)"
     conclusion_p_valor = f"Dado que el valor p ({valor_p:.4f}) es menor que el nivel de significancia α ({alpha}), rechazamos la hipótesis nula."
else:
     decision_p_valor = "No rechazar H0 (usando valor p)"
     conclusion_p_valor = f"Dado que el valor p ({valor_p:.4f}) es mayor que el nivel de significancia α ({alpha}), no rechazamos la hipótesis nula."

print(f"   Decisión Estadística (Criterio del Valor Crítico): {decision_estadistica}")
print(f"   Conclusión (Criterio del Valor Crítico): {conclusion_critica}")
print(f"   Decisión Estadística (Criterio del Valor p): {decision_p_valor}")
print(f"   Conclusión (Criterio del Valor p): {conclusion_p_valor}")

print(f"\n   Conclusión General en Términos Prácticos:")
if valor_p <= alpha:
     print(f"   Al nivel de significancia de {alpha*100}%, hay suficiente evidencia estadística para concluir que la verdadera varianza en el grosor de las piezas metálicas es significativamente menor a 0.5 mm².")
else:
     print(f"   Al nivel de significancia de {alpha*100}%, no hay suficiente evidencia estadística para concluir que la verdadera varianza en el grosor de las piezas metálicas es significativamente menor a 0.5 mm². No podemos afirmar con esta evidencia que la varianza poblacional sea menor que la garantizada.")

print(f"\n   Significado del Valor p:")
print(f"   El valor p de {valor_p:.4f} es la probabilidad de obtener una varianza muestral tan baja como 0.3 (o más baja) si la verdadera varianza poblacional fuera realmente 0.5 mm². Un valor p pequeño (< α) indica que el resultado observado es improbable bajo la hipótesis nula, lo que nos lleva a rechazarla.")
```
